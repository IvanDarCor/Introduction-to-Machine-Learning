#Redes Neuronales Convolucionales

El modelo es el que guarda el aprendizaje

$\hat{\mathbf{y}}=f\left(\mathbf{w}^T \mathbf{x}+b\right)$


$
\mathbf{w}$ representa el conocimiento adquirido

$
\mathbf{x}$ es el conjunto de muestras

$
\hat{\mathbf{y}} $ es la salida de la red


En clasificación multiclase se obtiene un vetor con la probabilidad de pertenecer a cada clase

$$
\hat{\mathbf{y}_n}=\left[\begin{array}{llll}
P\left(0 \mid \mathbf{x}_{\mathbf{n}}\right) & P\left(1 \mid \mathbf{x}_{\mathbf{n}}\right) & \cdots & P\left(C \mid \mathbf{x}_{\mathbf{n}}\right)
\end{array}\right]
$$



El modelo para una capa densa es:

$$
\begin{aligned}
\hat{\mathbf{Y}} &=f\left(\mathbf{X} \mathbf{W}^T+\mathbf{b}\right) \\
&=f\left(\left[\begin{array}{cccc}
x_{11} & x_{12} & \cdots & x_{1 p} \\
x_{21} & x_{22} & \cdots & x_{2 p} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n 1} & x_{n 2} & \cdots & x_{n p}
\end{array}\right]\left[\begin{array}{cccc}
w_{11} & w_{21} & \cdots & w_{q 1} \\
w_{12} & x_{22} & \cdots & w_{q 2} \\
\vdots & \vdots & \ddots & \vdots \\
w_{1 p} & w_{2 p} & \cdots & w_{q p}
\end{array}\right]+\left[\begin{array}{c}
b_1 \\
b_2 \\
\vdots \\
b_q
\end{array}\right]\right) \\
&=\left[\begin{array}{cccc}
P\left(0 \mid \mathbf{x}_1\right) & P\left(1 \mid \mathbf{x}_1\right) & \cdots & P\left(C \mid \mathbf{x}_1\right) \\
P\left(0 \mid \mathbf{x}_2\right) & P\left(1 \mid \mathbf{x}_2\right) & \cdots & P\left(C \mid \mathbf{x}_2\right) \\
\vdots & \vdots & \ddots & \vdots \\
P\left(0 \mid \mathbf{x}_{\mathbf{n}}\right) & P\left(1 \mid \mathbf{x}_{\mathbf{n}}\right) & \cdots & P\left(C \mid \mathbf{x}_{\mathbf{n}}\right)
\end{array}\right]
\end{aligned}
$$

El cuerpo de la neurona se encarga de realizar proceso sobre las señales de entrada para generar la salida, las funciones de activación imitan ese proceso, existen varios tipos que se utilizan de acuerdo con la tarea a resolver.

<img src="https://drive.google.com/uc?export=view&id=1Kmk_qfLlcqn9ijGWXRmuujF1MP19uUsL" width="50%" height="35%"/>

Al modelo se le especifica lo que debe aprender mediante las funciones de costo, dependiendo del tipo de aprendizaje y tarea que se desee desempeñar existen diferentes tipos de funciones de costo. En este caso son de interés los modelos de aprendizaje supervisados.
La función de perdida Permite evaluar el proceso de aprendizaje en una sola muestra $x_n$ del conjunto de entrenamiento.
La función de costo permite evaluar el proceso de aprendizaje en todo el conjunto de entrenamiento X.
La función de perdida más comúnmente utilizada para tareas de clasificación es la Entropía Cruzada.

$$
H(\mathbf{y}, \hat{\mathbf{y}})=-\langle\mathbf{y} \log (\hat{\mathbf{y}})\rangle
$$

<img src="https://drive.google.com/uc?export=view&id=19yXuj0PJhYR1S-yq_1FMqR-IQBq14PsK" width="50%" height="35%"/>

El objetivo es minimizar el valor de costo, así asegurar que el modelo desempeña bien la tarea sobre el conjunto de entrenamiento, pero se puede presentar sobreajuste y el modelo memorizaría los datos, para esto se utiliza un conjunto de prueba para saber si el modelo efectivamente aprendió los patrones relevantes para la clasificación

$$
C(\mathbf{Y}, \hat{\mathbf{Y}})=\frac{1}{N} \sum_{n=1}^N H\left(\mathbf{y}_n, \hat{\mathbf{y}}_n\right)
$$


Si escribimos $\hat{\mathbf{y}}_n$ en términos de la salida de una capa densa:

$$
C(\mathbf{X}, \mathbf{Y} ; \mathbf{W}, \mathbf{b})=\frac{1}{N} \sum_{n=1}^N H\left(\mathbf{y}_n, f\left(\mathbf{x}_n \mathbf{W}^T+\mathbf{b}\right)\right)
$$

Si escribimos


$\theta=\{\mathbf{W}, \mathbf{b}\}$

se tendría el siguiente problema de optimización.


$
\theta^*=\underset{\theta}{\arg \min } C(\mathbf{Y}, \hat{\mathbf{Y}})
$

Y encontrar el mínimo valor corresponde a la adquisición de aprendizaje.

La combinación lineal que realiza cada neurona sobre sus entradas, se ve desde el punto de
vista geométrico como un hiperplano con vector normal w.
Capas ocultas se encargan de mapear los datos de un espacio a otro, buscando que estos sean linealmente separables para que la capa de salida pueda operar sobre ellos.

<img src="https://drive.google.com/uc?export=view&id=1V-n-aqfnTr2PaIa79Z3_SL8ShNPm2PON" width="50%" height="35%"/>

El modelo de una red neuronal esta dado por la siguiente expresión.

$$
\hat{\mathbf{Y}}=\left(f_L \circ f_{L-1} \circ \cdots \circ f_1\right)(\mathbf{X})
$$
donde $L$ es el numero de capas ocultas, $f=f\left(\mathbf{X} \mathbf{W}^T+\mathbf{b}\right)$ es el modelo de una capa densa, $\mathbf{X}$ es el conjunto de datos $\mathbf{y} \mathbf{Y}$ es el conjunto de etiquetas.


<img src="https://drive.google.com/uc?export=view&id=1g032hBdEUpwSAB7S-2bK9TD94Hok29Ep" width="50%" height="35%"/>


**Metricas de evaluación del modelo:**


$\text { Exactitud }=\frac{M}{N} $

M,N son el número de muestras clasificadas correctamente y el total de muestras respectivamente.

$\text { presición }=\frac{\mathrm{VP}}{\mathrm{VP}+\mathrm{FP}}$

Mide la proporción de prediciones positivas que si pertenecen a esta clase.

$\text { exhaustividad }=\frac{\mathrm{VP}}{\mathrm{VP}+\mathrm{FN}}$

Mide la proporción de predicciones positivas que sí pertenecen a esta clase

$F_1=2 * \frac{\text { precisión } \times \text { exhaustividad }}{\text { precisión+exhaustividad }}$

Es una métrica que se utiliza para considerar al mismo tiempo tanto la precisión como la exhaustividad, con el objetivo de obtener el mejor valor de ambas métricas en conjunto.

$F_\beta=\left(1+\beta^2\right) * \frac{\text { precisión } \times \text { exhaustividad }}{\beta^2 \text { precisión }+\text { exhaustividad }}$

Es una version mejorada del F1 score, que permite dar mayor importancia a la presición o la exhaustividad dependiendo de la aplicación a traves del parámetro $\beta$

```python
import sys
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Flatten, Dense, Activation,Convolution2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator ,img_to_array,load_img
from tensorflow.keras import backend as K

from keras.preprocessing import image

from tensorflow.keras.preprocessing import image

from tensorflow.keras.utils import load_img,img_to_array
from keras.models import load_model

from tensorflow.keras.models import load_model
import argparse
import pickle
import cv2

from tensorflow.keras.applications.imagenet_utils import decode_predictions

import logging, os
import sys

import math
import operator
from pathlib import Path
```

```python
K.clear_session()

data_entrenamiento = '/content/drive/MyDrive/coffe berry/Entrenamiento'
data_validacion = '/content/drive/MyDrive/coffe berry/Validacion'


epocas=10
longitud, altura = 150, 150
batch_size = 1  # numero de imagenes en cada paso
pasos = 164
validation_steps = 152

#pasos = 10 #numero de veces que se procesa la informacion en cada epoca
#validation_steps = 10 #

filtrosConv1 = 32
filtrosConv2 = 64
tamano_filtro1 = (3, 3)
tamano_filtro2 = (2, 2)
tamano_pool = (2, 2)
clases = 5
lr = 0.0004

opt = tf.keras.optimizers.Adam(learning_rate=lr)

##Preparamos nuestras imagenes

entrenamiento_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1. / 255)

entrenamiento_generador = entrenamiento_datagen.flow_from_directory(
    data_entrenamiento,
    target_size=(altura, longitud),
    batch_size=batch_size,
    class_mode='categorical')

validacion_generador = test_datagen.flow_from_directory(
    data_validacion,
    target_size=(altura, longitud),
    batch_size=batch_size,
    class_mode='categorical')



cnn=tf.keras.Sequential()
cnn.add(tf.keras.layers.Conv2D(filters=64,padding='same',strides=2,kernel_size=3,activation='relu',input_shape=(longitud,altura,3)))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn.add(tf.keras.layers.Conv2D(filters=32,padding='same',strides=2,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))

cnn.add(tf.keras.layers.Conv2D(filters=32,padding='same',strides=2,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2))

cnn.add(tf.keras.layers.Flatten())
cnn.add(tf.keras.layers.Dense(5,activation='softmax'))  #5 es el numero de clases


cnn.compile(optimizer=tf.keras.optimizers.Adam(),loss='categorical_crossentropy',metrics=['accuracy'])

#pasos = len(data_entrenamiento)//batch_size

#validation_steps = len(data_validacion)//batch_size # if you have validation data

cnn.fit(
    entrenamiento_generador,
    steps_per_epoch=pasos,
    epochs=epocas,
    validation_data=validacion_generador,
    validation_steps=validation_steps)
```
